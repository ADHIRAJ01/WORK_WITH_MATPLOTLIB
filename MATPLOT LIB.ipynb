# ADDING STYLE TO GRAPH

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")

x1=[2,5,9]
y1=[7,2,5]

x2=[1,3,7]
y2=[8,4,6]

plt.plot(x1,y1,'g',label='line one',linewidth=5)  #giving width and color to line 
plt.plot(x2,y2,'r',label='two',linewidth=5)

plt.title('DESIGNER GRAPH')
plt.xlabel('X AXIS')
plt.ylabel('Y LABEL')

plt.legend()   # shows line one and two at right top corner

#plt.grid(False)
plt.grid(True,color='y')  # color of grid lines and grid

plt.show()
```

# UNIFORM VALUES IN BIG DATA PLOTTING

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 100000)

print(x)

plt.hist(x, 100)
plt.show()
```
